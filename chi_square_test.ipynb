# **Chi-Square Test for Independence**

* The Chi-Square Test of Independence determines whether there is an association between **categorical variable**s, i.e., whether the variables are independent or related.

* **Example**: Suppose we want to know whether or not gender is associated with political party preference. We take a  random sample of 1000 voters and survey them on their political party preference. 
* This survey utilizes a **contingency table** to analyze the data:

image.png
* A contingency table (also known as a cross-tabulation, crosstab, or two-way table) is an arrangement in which data is classified according to two categorical variables. 

* The categories for one variable appear in the rows, and the categories for the other variable appear in columns.
